```python
import numpy as np
import copy
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.patches import FancyArrowPatch
from numpy.typing import NDArray
```

## Auxiliary functions

```python
def scale(polygon, factor):
    points = []
    for x, y in polygon:
        new_x = x*factor
        new_y = y*factor
        points.append([new_x, new_y])
    return points

def angle_add(angle1: NDArray[np.float64], angle2: NDArray[np.float64]):
    cos1, sin1 = angle1
    cos2, sin2 = angle2
    return np.array([cos1*cos2 - sin1*sin2, cos1*sin2 + cos2*sin1])

def angle_sub(angle1: NDArray[np.float64], angle2: NDArray[np.float64]):
    # angle1 - angle2. Same as add, but sin2 swaps sign
    cos1, sin1 = angle1
    cos2, sin2 = angle2
    return np.array([cos1*cos2 + sin1*sin2, -cos1*sin2 + cos2*sin1])

def vector_sub(v1, v2):
    return np.array([v1[0] - v2[0], v1[1] - v2[1]])
    
def remove_frame(ax):
    for spine in ax.spines.values():
        spine.set_visible(False)
    
    ax.tick_params(
        left=False, right=False, 
        top=False, bottom=False, 
        labelleft=False, labelbottom=False)

def add_grid(ax, width, color):
    for i in range(1,12):
        ax.plot([0, 12], [i,i], lw = width, color=color)
        ax.plot([i,i], [0, 12], lw = width, color=color)

def plot_poly(ax, poly, lw, alpha, add_arrows = False):

    points = [v.position for v in poly.vertices]
    polygon = Polygon(points, closed=True, facecolor=poly.color, edgecolor='k', lw = 0, alpha = alpha)
    ax.add_patch(polygon)

    xs = [p[0] for p in points]
    ys = [p[1] for p in points]
    
    xs.append(xs[0])
    ys.append(ys[0])

    ax.plot(xs, ys, lw = lw, color = 'k')

    if add_arrows:
        for i in range(len(points)):
            x, y = xs[i], ys[i]
            end_x, end_y = xs[i+1], ys[i+1]
            mid_x, mid_y = (end_x + x)*0.5, (end_y + y)*0.5
            ax.annotate('', xy=(mid_x, mid_y), xytext=(x, y), arrowprops=dict(arrowstyle='-|>', color='k', mutation_scale=30))
```

```python
def place_poly_in_vertex(poly_index, probing_vertex_index, anchor_vertex):
        
    poly = polys[poly_index]
    poly.reset()
    probing_vertex = poly.vertices[probing_vertex_index]
    # print(probing_vertex.position, probing_vertex.angle_to, probing_vertex.angle_from)
    
    angle_dif = angle_sub(probing_vertex.angle_to, anchor_vertex.angle_from)
    dr = vector_sub(anchor_vertex.position, probing_vertex.position)
    
    poly.translate(dr)
    poly.rotate(angle_dif, anchor_vertex.position)

    return poly
```

```python
class Vertex:
    def __init__(self, p, prev, nex):
        self.position   =  np.array([p[0], p[1]], dtype=float)
        self.angle_to   =  np.array([nex[0] - p[0], nex[1] - p[1]], dtype=float)
        self.angle_from = -np.array([p[0] - prev[0], p[1] - prev[1]], dtype=float)

        self.angle_to   /= np.linalg.norm(self.angle_to)
        self.angle_from /= np.linalg.norm(self.angle_from)

class MyPolygon:
    def __init__(self, points):
        self.vertices = []
        

        N = len(points)
        
        for i in range(N):
            current = points[i]
            pre = points[(i-1+N)%N]
            nex = points[(i+1)%N]
            self.vertices.append(Vertex(current, pre, nex))

        self.vertices_immutable = copy.deepcopy(self.vertices)

    def reset(self):
        self.vertices = copy.deepcopy(self.vertices_immutable)
        
    def set_color(self, color):
        self.color = color

    def translate(self, vector):
        for v in self.vertices:
            v.position += vector

    def move_to(self, vector):
        dr = vector - self.vertices[0].position
        for v in self.vertices:
            v.position += dr
            
    """def translate_immutable(self, vector):
        for v, vi in zip(self.vertices, self.vertices_immutable):
            v.position = vi.position + vector"""

    def rotate(self, angle, rotation_center = np.zeros(2)):
        cos, sin = angle
        for v in self.vertices:
            x, y = v.position - rotation_center

            new_x =  x*cos + y*sin
            new_y = -x*sin + y*cos

            v.position = np.array([new_x, new_y]) + rotation_center
            v.angle_to = angle_sub(v.angle_to, angle)
            v.angle_from = angle_sub(v.angle_from, angle)
            
    """def rotate_immutable(self, angle, rotation_center = np.zeros(2)):
        cos, sin = angle
        for v, vi in zip(self.vertices, self.vertices_immutable):
            x, y = vi.position - rotation_center

            new_x =  x*cos + y*sin
            new_y = -x*sin + y*cos

            v.position = np.array([new_x, new_y]) + rotation_center
            v.angle_to = angle_sub(v.angle_to, angle)
            v.angle_from = angle_sub(v.angle_from, angle)"""

    def get_center(self):
        xc, yc = 0, 0
        N = len(self.vertices)
        for v in self.vertices:
            x, y = v.position
            
            xc += x
            yc += y
        
        xc /= N
        yc /= N
        return xc, yc


    def get_center_immutable(self):
        xc, yc = 0, 0
        N = len(self.vertices)
        for v in self.vertices_immutable:
            x, y = v.position
            
            xc += x
            yc += y
        
        xc /= N
        yc /= N
        return xc, yc

    def permute_position(self, n):
        positions = []
        N = len(self.vertices)
        for i in range(N):
            j = (i + n)%N
            positions.append(self.vertices[j].position)
        return positions
        
```

```python
frame = [[0, 0], [0, 12], [12, 12], [12, 0]]
poly1  = [[0,0],  [3,0],  [2,4]]
poly2  = [[3,0],  [3,6],  [2,4]]
poly3  = [[3,0],  [6,0],  [6,6], [4,8], [3,6]]
poly4  = [[0,0],  [4,8],  [2,10]]
poly5  = [[0,0],  [2,10], [0,12]]
poly6  = [[0,12], [4,8],  [6,12]]
poly7  = [[4,8],  [6,6],  [6,12]] 
poly8  = [[8,4],  [12,0], [9,6]]
poly9  = [[9,6],  [12,0], [12,6]]
poly10 = [[9,6],  [12,6], [12,8]]
poly11 = [[6,6],  [8,4],  [9,6], [6,12]]
poly12 = [[9,6],  [12,8], [12,12], [6,12]]
poly13 = [[6,0],  [12,0], [8,4]]
poly14 = [[6,0],  [8,4],  [6,6]]

poly_array = [poly1, poly2, poly3, poly4, poly5, poly6, poly7, 
         poly8, poly9, poly10, poly11, poly12, poly13, poly14]

cmap = plt.get_cmap('tab20')
colors = [cmap(i) for i in range(len(poly_array))]

polys = []
for points, color in zip(poly_array, colors):
    polygon = MyPolygon(points)
    polygon.set_color(color)
    polys.append(polygon)

frame = MyPolygon(frame)
frame.set_color("w")
```

## Recursive

```python
def branch1a(position):
    frame.move_to(position)
    poly = place_poly_in_vertex(12, 1, frame.vertices[0])
    plot_poly(ax, frame, lw, 1.0)
    plot_poly(ax, poly, 0.0, 1.0)

def branch1a2a(position):
    frame.move_to(position)
    poly = place_poly_in_vertex(12, 1, frame.vertices[0])
    poly2 = place_poly_in_vertex(0, 2, poly.vertices[1])
    plot_poly(ax, frame, lw, 1.0)
    plot_poly(ax, poly, 0.0, 1.0)
    plot_poly(ax, poly2, 0.0, 1.0)

def branch1a2b(position):
    frame.move_to(position)
    poly = place_poly_in_vertex(12, 1, frame.vertices[0])
    poly2 = place_poly_in_vertex(2, 1, frame.vertices[1])
    plot_poly(ax, frame, lw, 1.0)
    plot_poly(ax, poly, 0.0, 1.0)
    plot_poly(ax, poly2, 0.0, 1.0)


def branch1a2b3a(position):
    frame.move_to(position)
    poly = place_poly_in_vertex(12, 1, frame.vertices[0])
    poly2 = place_poly_in_vertex(2, 1, frame.vertices[1])
    poly3 = place_poly_in_vertex(0, 2, poly.vertices[1])
    plot_poly(ax, frame, lw, 1.0)
    plot_poly(ax, poly, 0.0, 1.0)
    plot_poly(ax, poly2, 0.0, 1.0)
    plot_poly(ax, poly3, 0.0, 1.0)

def branch1a2b3b(position):
    frame.move_to(position)
    poly = place_poly_in_vertex(12, 1, frame.vertices[0])
    poly1 = place_poly_in_vertex(2, 1, frame.vertices[1])
    poly2 = place_poly_in_vertex(5, 0, poly1.vertices[0])
    plot_poly(ax, frame, lw, 1.0)
    plot_poly(ax, poly, 0.0, 1.0)
    plot_poly(ax, poly1, 0.0, 1.0)
    plot_poly(ax, poly2, 0.0, 1.0)
```

```python
def points_to_coords(points):
    xs = [r[0] for r in points]
    ys = [r[1] for r in points]
    return xs, ys

def plot_lines(ax, points, lw, color):
    xs, ys = points_to_coords(points)

    ax.plot(xs, ys, color = color, linewidth = lw)
    
    for i in range(len(points)-1):
        x, y = xs[i], ys[i]
        end_x, end_y = xs[i+1], ys[i+1]
        mid_x, mid_y = (end_x + x)*0.5, (end_y + y)*0.5
        ax.annotate('', xy=(mid_x, mid_y), xytext=(x, y), arrowprops=dict(arrowstyle='-|>', color=color, mutation_scale=30))
        
        
    
points  = [[-1.0, 0.0], [0.5, 0.0], [2.0, 0.0], [3.5, 0.0], [5.0, 0.0], [6.5, 0.0]]
points1 = [[0.0, 1.0], [0.5, 0.0], [1.0, 1.0]]
points2 = [[0.5, -1.0], [1.0, 0.0], [1.5, -1.0]]
points3 = [[2.0, 1.0], [2.0, 0.0], [2.5, -1.0]]
points4 = [[3.2, 0.0], [4.0, 0.0], [4.5, -1.0]]
points5 = [[4.7, 0.0], [5.5, 0.0], [6.0, 1.0]]

fig, ax = plt.subplots(1,1, figsize=(10,2))
ax.set_xlim([-1, 10])
ax.set_ylim([-1.1, 1.1])
remove_frame(ax)

lw = 3.0
plot_lines(ax, points[::-1],  1.5, "k")
plot_lines(ax, points1, lw, "C3")
plot_lines(ax, points2, lw, "C0")
plot_lines(ax, points3, lw, "C3")
plot_lines(ax, points4, lw, "C0")
plot_lines(ax, points5, lw, "C3")

plt.show()
```

```python

def draw_wedge(ax, origin, angle_start, angle_end, color):

        
    u1 = angle_start / np.linalg.norm(angle_start)
    u2 = angle_end / np.linalg.norm(angle_end)
    

    # Compute angles of the vectors w.r.t. x-axis (in degrees)
    angle1 = np.degrees(np.arctan2(u1[1], u1[0]))
    angle2 = np.degrees(np.arctan2(u2[1], u2[0]))
    
    # Compute start angle and sweep angle (ensure counter-clockwise direction)
    start_angle = angle1
    sweep_angle = (angle2 - angle1) % 360
    
    # Draw a filled circular sector using Wedge
    arc_radius = 0.7
    wedge = patches.Wedge(center=origin, r=arc_radius,
                          theta1=start_angle, theta2=start_angle + sweep_angle,
                          facecolor=color, alpha=0.3)
    
    ax.add_patch(wedge)
    
```

```python

fig, ax = plt.subplots(1,1, figsize=(8,4))
ax.set_xlim([-2, 10])
ax.set_ylim([-2, 2])
remove_frame(ax)

origin = np.array([0,0])

angle_fr = np.array([1.5,  0.0])
angle_to = np.array([0.0, -1.5])
draw_wedge(ax, origin, angle_to, angle_fr, 'g')
plot_lines(ax, [angle_fr+origin] + [origin] + [angle_to+origin], 2.0, 'k')

angle_fr = np.array([1.0,  1.0])
angle_to = np.array([1.0, -1.0])
draw_wedge(ax, origin, angle_to, angle_fr, 'C3')
plot_lines(ax, [angle_fr+origin] + [origin] + [angle_to+origin], 2.0, 'k')



origin = np.array([3,0])

angle_fr = np.array([-1.0, -1.0])
angle_to = np.array([1.2,  -0.3])
draw_wedge(ax, origin, angle_to, angle_fr, 'g')
plot_lines(ax, [angle_fr+origin] + [origin] + [angle_to+origin], 2.0, 'k')

angle_fr = np.array([1.0,  -1.0])
angle_to = np.array([0.0, -1.0])
draw_wedge(ax, origin, angle_to, angle_fr, 'C0')
plot_lines(ax, [angle_fr+origin] + [origin] + [angle_to+origin], 2.0, 'k')


origin = np.array([6,0])

angle_fr = np.array([-1.0, -1.0])
angle_to = np.array([1.2,  -0.3])
draw_wedge(ax, origin, angle_to, angle_fr, 'g')
plot_lines(ax, [angle_fr+origin] + [origin] + [angle_to+origin], 2.0, 'k')

angle_fr = np.array([-1.0,  1.0])
angle_to = np.array([0.0, 1.0])
draw_wedge(ax, origin, angle_to, angle_fr, 'C3')
plot_lines(ax, [angle_fr+origin] + [origin] + [angle_to+origin], 2.0, 'k')



plt.show()
```
